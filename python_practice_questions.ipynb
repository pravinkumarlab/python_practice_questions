# **1. Odd or Even Sum**

Problem Statement:
Given 2 numbers N and M add both the numbers and check whether the sum is odd or even.

```python
N,M = map(int, input().split())
print("even" if (N+M)%2==0 else "odd")
```

# **2. Prime Number Check**

Problem Statement:
Given a number N, check whether it is prime or not. Print 'yes' if it is prime else print 'no'.


```python
N = int(input())
if N>1:
  for i in range(2,int(N**0.5)+1):
    if N % i == 0:
      print("no")
      break
  else:
    print("yes")
else:
  print("no")
```

# **4. Next Greater Power of 2**
Question:
Given a number N, find its next immediate greater power of 2(i.e 2^1, 2^2, 2^3...).

```python
N = int(input())
power = 1
while power <= N:
    power *= 2
print(power)
```

# **5. Palindrome Check**
Problem Statement:
Given a string S, print 'yes' if it is a palindrome or 'no' if it is not a palindrome.

```python
S = input()
print("yes" if S==S[::-1] else "no")
```

# **6. Vowel Check in String**
Question:
Given a string S, print 'yes' if it has a vowel in it else print 'no'.

```python
S = input()
vowels = "aeiouAEIOU"
print("yes" if any(ch in vowels for ch in S) else "no")
```

# **8. Round to Nearest Greater Integer**

# **9. String Middle Element Modification**
Question:
Given a string S, print it after changing the middle element to * (if the length of the string is even, change the 2 middle elements to *).

```python
S = input()
n = len(S)
if n % 2 != 0:
    mid = n // 2
    S = S[:mid] + "*" + S[mid+1:]
else:
    mid1 = n // 2 - 1
    mid2 = n // 2 + 1
    S = S[:mid1] + "**" + S[mid2:]
print(S)
```

# **10. Check if Number is Composite**
Problem Statement:
Given a number N, print 'yes' if it is composite else print 'no'.

```python
N = int(input())

if N > 1:
    for i in range(2, int(N**0.5) + 1):
        if N % i == 0:
            print("yes")
            break
    else:
        print("no")
else:
    print("no")
```

## **4. Pyramid Pattern**

**Problem Statement:**
Given an integer N, print a pyramid pattern of asterisks.


```python
def pyramid():
    n = int(input())
    for i in range(1, n + 1):
        spaces = "  " * (n - i)
        stars = " ".join(["*"] * (2*i - 1))
        print(spaces + stars)
pyramid()
```
